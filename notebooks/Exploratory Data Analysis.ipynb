```python
# function to assign new values to field 

def a(row):
    if row['field_cd'] == 2.0 or row['field_cd'] == 4.0 or row['field_cd'] == 5.0 or row['field_cd'] == 10.0 or row['field_cd'] == 17.0:
        val = 1
    elif row['field_cd'] == 1 or row['field_cd'] == 1 or row['field_cd'] == 3 or row['field_cd'] == 7 or row['field_cd'] == 8 or row['field_cd'] == 9 or row['field_cd'] == 11 or row['field_cd'] == 13 or row['field_cd'] == 16:
        val = 2
    elif row['field_cd'] == 6 or row['field_cd'] == 14 or row['field_cd'] == 15:
        val = 3
    else:
        val = 4
    return val
```

**Comparing attributes and changes before vs after dating event**

```python
# What you look for in opposite sex? (100 point allocation)

you_look = df[['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1',
              'attr1_2', 'sinc1_2', 'intel1_2', 'fun1_2', 'amb1_2', 'shar1_2']]

_ = sns.heatmap(you_look.corr(), linewidth=.5, vmin=-1, center=0, cmap="PiYG")
```

Attractiveness has a weak to medium negative correlation with the rest of the attributes. This indicates that as attractiveness increased in importance, the rest of the attributes decreased and vice versa. 

There's a high but not perfect positive correlation between ratings on what participants look for in their partners before and after the dating event. It appears that their opinions changed slightly during this time period. It may not be a direct result of the dating event itself, but it's probably that there was some influence. 

```python
# What you think MOST of your fellow men/women look for in the opposite sex? (100 point allocation)

other_same_look = df[['attr4_1', 'sinc4_1', 'intel4_1', 'fun4_1', 'amb4_1', 'shar4_1',
                     'attr4_2', 'sinc4_2', 'intel4_2', 'fun4_2', 'amb4_2', 'shar4_2']]

_ = sns.heatmap(other_same_look.corr(), linewidth=.5, vmin=-1, center=0, cmap="PiYG")
```

Attractiveness has a much weaker negative correlation with the rest of the attributes in the case that participants are asked to rate what others of the same gender look for in their partners. The rest of the attributes generally have a weak to medium positive correlation with each other. This indicates that participants were probably more neutral and less particular about what they thought others of the same gender valued. 

```python
# What do you think the opposite sex looks for in a date? (100 point allocation)

other_look = df[['attr2_1', 'sinc2_1', 'intel2_1', 'fun2_1', 'amb2_1', 'shar2_1',
                'attr2_2', 'sinc2_2', 'intel2_2', 'fun2_2', 'amb2_2', 'shar2_2']]

_ = sns.heatmap(other_look.corr(), linewidth=.5, vmin=-1, center=0, cmap="PiYG")
```

Attractiveness has a weak to strong negative correlation with the rest of the attributes. It's more extreme than in the case of what participants said they looked for themselves.
