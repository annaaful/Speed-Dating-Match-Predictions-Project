## Column Keys 

iid: unique subject number, group(wave id gender) <br>
id: subject number within wave <br>
gender: female=0, male=1 <br>
idg: 	subject number within gender, group(id gender) <br>
round: number of people that met that wave <br>
order: the number of date that night when met partner  <br>
partner: 	partner’s id number the night of event <br>
pid: 		partner’s iid number <br>
match:		1=yes, 0=no <br>
int_corr:	correlation between participant’s and partner’s ratings of interests in Time 1 <br>
samerace:  	participant and the partner were the same race. 1= yes, 0=no <br>
age_o:		age of partner <br>
race_o:		race of partner <br>
pf_o_att: 	partner’s stated preference at Time 1 (attr1_1) for all 6 attributes <br>
dec_o: 		decision of partner the night of event <br>
attr_o: 		rating by partner the night of the event, for all 6 attributes <br>
